# 1. Data Preparation

```python
import pandas as pd

file_path = r"C:\Users\hp\OneDrive\Documents\P PROJECTS\h\Police_Bulk_Data_2014_20241116.csv"
df = pd.read_csv(file_path)

print("First few rows of the dataset:")
print(df.head())

print("\nSummary information:")
print(df.info())

# Checking for missing values percentage per column
missing_percentage = (df.isnull().sum() / len(df)) * 100
print("\nMissing values percentage per column:")
print(missing_percentage)

# Dropping columns with more than 50% missing values
threshold = 50  # Percentage
columns_to_drop = missing_percentage[missing_percentage > threshold].index
print(f"\nColumns to drop (more than {threshold}% missing values): {columns_to_drop.tolist()}")
df = df.drop(columns=columns_to_drop)

# Imputing missing values in remaining columns (example: fill with mean for numeric columns)
for col in df.select_dtypes(include=['float64', 'int64']):
    df[col].fillna(df[col].mean(), inplace=True)

# Dropping rows with missing values in non-numeric columns
df = df.dropna()

# Displaying the shape of the cleaned dataset
print("\nShape of the dataset after cleaning:")
print(df.shape)

cleaned_file_path = r"C:\Users\hp\OneDrive\Documents\P PROJECTS\h\Cleaned_Police_Bulk_Data.csv"
df.to_csv(cleaned_file_path, index=False)
print(f"\nCleaned dataset saved at {cleaned_file_path}")

```

# 2. EDA
    

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

file_path = r"C:\Users\hp\OneDrive\Documents\P PROJECTS\h\Cleaned_Police_Bulk_Data.csv"
df = pd.read_csv(file_path)

# Convert 'offensedate' to datetime format and extract the day of the week
df['offensedate'] = pd.to_datetime(df['offensedate'], errors='coerce')
df['offenseday'] = df['offensedate'].dt.day_name()

# 1. Univariate Analysis: Distribution of offense start times
plt.figure(figsize=(10, 6))
sns.histplot(pd.to_datetime(df['offensestarttime'], format='%H:%M:%S', errors='coerce').dt.hour, 
             bins=24, kde=False, color='blue')
plt.title('Distribution of Offense Start Times', fontsize=14)
plt.xlabel('Hour of the Day')
plt.ylabel('Frequency')
plt.show()

# 2. Bivariate Analysis: Relationship between day of the week and offenses
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='offenseday', order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Number of Offenses by Day of the Week', fontsize=14)
plt.xlabel('Day of the Week')
plt.ylabel('Number of Offenses')
plt.show()

# 3. Multivariate Analysis: Correlation heatmap of numerical columns
plt.figure(figsize=(12, 8))
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df[numeric_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Variables', fontsize=14)
plt.show()

# 4. Insights: Display basic statistics
print("\nSummary Statistics for Numerical Columns:")
print(df.describe())

# Save basic stats to a file for reference
stats_file_path = r"C:\Users\hp\OneDrive\Documents\P PROJECTS\h\EDA_Summary.csv"
df.describe().to_csv(stats_file_path)
print(f"\nSummary statistics saved to: {stats_file_path}")
```

### Key Insights:

- Distribution of Offense Times:The above histogram of offensestarttime suggests that offenses are most likely to occur at those hours, perhaps early evening, when people are more outdoors or at places of congregation.

- Offenses by Day of the Week:The count plot indicates that most offenses happen over the weekend, most especially on Saturdays, and could be connected to the high rate of public activities.
